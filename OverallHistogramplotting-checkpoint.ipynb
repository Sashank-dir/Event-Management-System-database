```python
import numpy as np
import matplotlib.pyplot as plt
import re

# Defining the number of bins for histograms
bins = 40

# Defining the file paths for each query and each database 
files = {
    "Query 1": {
        "MongoDB": {"250K": "results_query1_mongodb_event250k.txt", "500K": "results_query1_mongodb_event500k.txt", "750K": "results_query1_mongodb_event750k.txt", "1M": "results_query1_mongodb_event1M.txt"},
        "Neo4j": {"250K": "results_query1_neo4j_event250k.txt", "500K": "results_query1_neo4j_event500k.txt", "750K": "results_query1_neo4j_event750k.txt", "1M": "results_query1_neo4j_event1M.txt"},
        "Redis": {"250K": "results_query1_redis_event250k.txt", "500K": "results_query1_redis_event500k.txt", "750K": "results_query1_redis_event750k.txt", "1M": "results_query1_redis_event1M.txt"},
        "MySQL": {"250K": "results_query1_mysql_event250k.txt", "500K": "results_query1_mysql_event500k.txt", "750K": "results_query1_mysql_event750k.txt", "1M": "results_query1_mysql_event1M.txt"},
        "Cassandra": {"250K": "results_query1_cassandra_event250k.txt", "500K": "results_query1_cassandra_event500k.txt", "750K": "results_query1_cassandra_event750k.txt", "1M": "results_query1_cassandra_event1M.txt"}
    },
    "Query 2": {
        "MongoDB": {"250K": "results_query2_mongodb_event250k.txt", "500K": "results_query2_mongodb_event500k.txt", "750K": "results_query2_mongodb_event750k.txt", "1M": "results_query2_mongodb_event1M.txt"},
        "Neo4j": {"250K": "results_query2_neo4j_event250k.txt", "500K": "results_query2_neo4j_event500k.txt", "750K": "results_query2_neo4j_event750k.txt", "1M": "results_query2_neo4j_event1M.txt"},
        "Redis": {"250K": "results_query2_redis_event250k.txt", "500K": "results_query2_redis_event500k.txt", "750K": "results_query2_redis_event750k.txt", "1M": "results_query2_redis_event1M.txt"},
        "MySQL": {"250K": "results_query2_mysql_event250k.txt", "500K": "results_query2_mysql_event500k.txt", "750K": "results_query2_mysql_event750k.txt", "1M": "results_query2_mysql_event1M.txt"},
        "Cassandra": {"250K": "results_query2_cassandra_event250k.txt", "500K": "results_query2_cassandra_event500k.txt", "750K": "results_query2_cassandra_event750k.txt", "1M": "results_query2_cassandra_event1M.txt"}
    },
    "Query 3": {
        "MongoDB": {"250K": "results_query3_mongodb_event250k.txt", "500K": "results_query3_mongodb_event500k.txt", "750K": "results_query3_mongodb_event750k.txt", "1M": "results_query3_mongodb_event1M.txt"},
        "Neo4j": {"250K": "results_query3_neo4j_event250k.txt", "500K": "results_query3_neo4j_event500k.txt", "750K": "results_query3_neo4j_event750k.txt", "1M": "results_query3_neo4j_event1M.txt"},
        "Redis": {"250K": "results_query3_redis_event250k.txt", "500K": "results_query3_redis_event500k.txt", "750K": "results_query3_redis_event750k.txt", "1M": "results_query3_redis_event1M.txt"},
        "MySQL": {"250K": "results_query3_mysql_event250k.txt", "500K": "results_query3_mysql_event500k.txt", "750K": "results_query3_mysql_event750k.txt", "1M": "results_query3_mysql_event1M.txt"},
        "Cassandra": {"250K": "results_query3_cassandra_event250k.txt", "500K": "results_query3_cassandra_event500k.txt", "750K": "results_query3_cassandra_event750k.txt", "1M": "results_query3_cassandra_event1M.txt"}
    },
    "Query 4": {
        "MongoDB": {"250K": "results_query4_mongodb_event250k.txt", "500K": "results_query4_mongodb_event500k.txt", "750K": "results_query4_mongodb_event750k.txt", "1M": "results_query4_mongodb_event1M.txt"},
        "Neo4j": {"250K": "results_query4_neo4j_event250k.txt", "500K": "results_query4_neo4j_event500k.txt", "750K": "results_query4_neo4j_event750k.txt", "1M": "results_query4_neo4j_event1M.txt"},
        "Redis": {"250K": "results_query4_redis_event250k.txt", "500K": "results_query4_redis_event500k.txt", "750K": "results_query4_redis_event750k.txt", "1M": "results_query4_redis_event1M.txt"},
        "MySQL": {"250K": "results_query4_mysql_event250k.txt", "500K": "results_query4_mysql_event500k.txt", "750K": "results_query4_mysql_event750k.txt", "1M": "results_query4_mysql_event1M.txt"},
        "Cassandra": {"250K": "results_query4_cassandra_event250k.txt", "500K": "results_query4_cassandra_event500k.txt", "750K": "results_query4_cassandra_event750k.txt", "1M": "results_query4_cassandra_event1M.txt"}
    }
}

# Function to load data by extracting floating-point numbers and convert to milliseconds
def load_execution_times(file_name):
    with open(file_name, 'r') as file:
        content = file.read()  # To read the entire file content as a string
    # Extract all floating-point numbers from the content
    times = re.findall(r'\d+\.\d+', content)  # Finds all matches of decimal numbers
    # Convert the extracted strings to floats and scale them to milliseconds (multiplying by 1000)
    return np.array(times, dtype=float) * 1000  # Convert seconds to milliseconds

# Dictionary to store execution times after parsing files
query_execution_times = []
for query_name, db_data in files.items():
    query_data = {}
    for db_name, dataset_files in db_data.items():
        dataset_data = {}
        for dataset_size, file_name in dataset_files.items():
            # Load and parse each file to extract numerical execution times in milliseconds
            dataset_data[dataset_size] = load_execution_times(file_name)
        query_data[db_name] = dataset_data
    query_execution_times.append(query_data)

# Dataset size labels for titles
dataset_sizes = ["250K", "500K", "750K", "1M"]

# Function to calculate the average of the subsequent 30 executions
def calculate_average_execution_times(execution_times):
    # Exclude the first execution time, take the next 30, and calculate the average
    return np.mean(execution_times[1:31])

# Iterate over each query to create histograms for all databases and dataset sizes
for query_index, query_data in enumerate(query_execution_times, start=1):
    # Adjust the subplot size for 4 rows and 5 columns (4 dataset sizes x 5 DBs)
    fig, axes = plt.subplots(4, 10, figsize=(20, 16))  # 4 rows and 10 columns
    fig.suptitle(f'Execution Times for Query {query_index} (in milliseconds)', fontsize=18)

    # Separate histograms for first execution and average of subsequent 30 executions
    for row, dataset_size in enumerate(dataset_sizes):
        for col, (db_name, db_data) in enumerate(query_data.items()):
            # Extract execution times for the first and average execution
            execution_times = db_data[dataset_size]
            first_execution_time = execution_times[0]
            # Calculate the average execution time for the next 30 executions
            average_execution_time = calculate_average_execution_times(execution_times)
            
            # Plot histogram for the first execution time (side by side)
            axes[row, 2*col].hist(first_execution_time, bins=bins, alpha=0.7, color='blue', edgecolor='black')
            axes[row, 2*col].set_title(f'{db_name} - {dataset_size} (First Exec)', fontsize=10)
            axes[row, 2*col].set_ylabel('Frequency', fontsize=8)
            axes[row, 2*col].tick_params(axis='both', labelsize=8)
            
            # Plot histogram for the average execution time (side by side)
            axes[row, 2*col + 1].hist(average_execution_time, bins=bins, alpha=0.7, color='green', edgecolor='black')
            axes[row, 2*col + 1].set_title(f'{db_name} - {dataset_size} (Avg Exec)', fontsize=10)
            axes[row, 2*col + 1].set_ylabel('Frequency', fontsize=8)
            axes[row, 2*col + 1].tick_params(axis='both', labelsize=8)

            # Add the "Execution Times" title below the histograms
            axes[row, 2*col].set_xlabel('Execution Time (ms)', fontsize=10)
            axes[row, 2*col + 1].set_xlabel('Execution Time (ms)', fontsize=10)

    # Adjust layout to prevent overlap and provide space for titles
    plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust the layout to fit within the figure area
    plt.show()
```
