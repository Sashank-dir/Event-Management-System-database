```python
import csv
import redis
import time

# Connect to your Redis instance
r = redis.Redis(host='localhost', port=6379, db=0)

# Define the dataset size and CSV filename
dataset_size = 250000

# Insert CSV data into Redis
with open('Downloads/event_management_dataset (1).csv', 'r') as file:
    csv_data = csv.reader(file)

    # Skip the header row if present
    next(csv_data)

    # Iterate over the rows and insert the data into Redis
    for row in csv_data:
        # Check if the row has the correct number of values
        if len(row) != 10:
            continue
        
        # Unpack the event management system fields
        event_name, venue, payment_status, seats_booked, ticket_price, attendee_name, attendee_email, attendee_phone, registration_date, event_date = row

        # Key for the event-attendee hash
        event_key = f'event:{event_name}:{attendee_name}'

        # Insert event-attendee data into Redis hash
        r.hset(event_key, 'Event Name', event_name)
        r.hset(event_key, 'Venue', venue)
        r.hset(event_key, 'Payment Status', payment_status)
        r.hset(event_key, 'Seats Booked', seats_booked)
        r.hset(event_key, 'Ticket Price', ticket_price)
        r.hset(event_key, 'Attendee Name', attendee_name)
        r.hset(event_key, 'Email', attendee_email)
        r.hset(event_key, 'Phone', attendee_phone)
        r.hset(event_key, 'Registration Date', registration_date)
        r.hset(event_key, 'Event Date', event_date)

        # Insert into sorted set with ticket price as the score
        sorted_set_key = 'events:ticket_price'
        r.zadd(sorted_set_key, {event_name: float(ticket_price)})

print(f"Data loaded into Redis for {dataset_size} events.")

# Benchmarking query performance on sorted set
query_times = []

# Query top 10 events by ticket price
for _ in range(30):
    start_rank = 0
    end_rank = 9
    start_time = time.time()
    results = r.zrange('events:ticket_price', start_rank, end_rank, withscores=True)
    end_time = time.time()
    execution_time = end_time - start_time
    query_times.append(execution_time)

# Write the execution times to a result file
with open(f'results_query2_redis_event250k.txt', 'w') as result_file:
    result_file.write(f"Query 2 execution times: {query_times}\n")

print("Query execution times saved to results file.")

# Close the Redis connection
r.close()

```
